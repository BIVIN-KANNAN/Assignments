```python
import pandas as pd
import numpy as np
import statistics as stat
```

```python
data = pd.read_csv('/content/StudentsPerformance.csv')
```

**1. Find out how many males and females participated in the test.**

```python
data.dtypes
```

```python
count_male = data['gender'].value_counts().get('male', 0)
count_male
```

```python
count_female = data['gender'].value_counts().get('female', 0)
count_female
```

**2. What do you think about the students' parental level of education?**

```python
parental_education_count = data['parental level of education'].value_counts()
parental_education_count
```

**3. Who scores the most on average for math, reading and writing based on**

**● Gender ● Test preparation course**

```python
avg_score_gender = data.groupby('gender').mean(numeric_only=True)[['math score', 'reading score', 'writing score']]
avg_score_course = data.groupby('test preparation course').mean(numeric_only=True)[['math score', 'reading score', 'writing score']]
print("\nAverage score based on Gender:")
print(avg_score_gender)

print("\nAverage score based on Course:")
print(avg_score_course)
```

**4. What do you think about the scoring variation for math, reading and writing based on
● Gender
● Test preparation course**

```python
score_variation_gender = data.groupby('gender').std(numeric_only=True)[['math score', 'reading score', 'writing score']]
score_variation_course = data.groupby('test preparation course').std(numeric_only=True)[['math score', 'reading score', 'writing score']]

print("\nScoring Variation based on Gender:")
print(score_variation_gender)

print("\nScoring Variation based on Test Preparation Course:")
print(score_variation_course)
```

**5. The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this.**

```python
top_25_percentile_math = np.percentile(data['math score'], 75)
data['bonus_points'] = np.where(data['math score'] >= top_25_percentile_math, 5, 0)
print("\nUpdated Dataset with Bonus Points:")
data.head()
```
